```python
import pandas as pd
import os
import glob
import warnings

# ÂøΩÁï• openpyxl Áõ∏ÂÖ≥ÁöÑË≠¶ÂëäÔºåÁ°Æ‰øùËæìÂá∫Ê∏ÖÊô∞
warnings.simplefilter(action='ignore', category=FutureWarning)

# 1. ÂÆö‰πâÊú¨Âú∞‰ªìÂ∫ìÁöÑË∑ØÂæÑ
# ÂÅáËÆæÊÇ®Âú®ËøêË°å git clone ÂêéÔºå‰ªìÂ∫ìÊñá‰ª∂Â§πÂ∞±Âú®ÂΩìÂâçÂ∑•‰ΩúÁõÆÂΩï‰∏ã
REPO_NAME = "Astraeusyataibei" 
LOCAL_REPO_PATH = os.path.join(os.getcwd(), REPO_NAME)

# 2. ÂàõÂª∫‰∏Ä‰∏™Â≠óÂÖ∏Êù•Â≠òÂÇ®ÊâÄÊúâÂä†ËΩΩÁöÑÊï∞ÊçÆ
data_frames = {}
text_contents = {}

print(f"Â∞ùËØï‰ªéÊú¨Âú∞ÁõÆÂΩïÂä†ËΩΩÊï∞ÊçÆ: {LOCAL_REPO_PATH}")

# 3. ÈÅçÂéÜ‰ªìÂ∫ì‰∏≠ÁöÑÊâÄÊúâÊñá‰ª∂
try:
    # glob ÊâæÂà∞ÊâÄÊúâ .xlsx Âíå .txt Êñá‰ª∂
    # ‰ΩøÁî® **/*.** Êù•ÈÄíÂΩíÊü•ÊâæÂ≠êÁõÆÂΩïÔºàÂ¶ÇÊûúÂ≠òÂú®ÔºâÔºå‰∏çËøáÊÇ®ÁöÑÊà™ÂõæÊòæÁ§∫Êñá‰ª∂ÈÉΩÂú®Ê†πÁõÆÂΩï
    all_files = glob.glob(os.path.join(LOCAL_REPO_PATH, '*.*'))

    if not all_files:
        print(f"‚ùå Êú™ÊâæÂà∞‰ªª‰ΩïÊñá‰ª∂„ÄÇËØ∑Á°ÆËÆ§ {LOCAL_REPO_PATH} ÁõÆÂΩïÂ≠òÂú®‰∏îÂåÖÂê´Êï∞ÊçÆÊñá‰ª∂„ÄÇ")
        print("ÊÇ®ÂèØËÉΩÈúÄË¶ÅÂÆâË£Ö openpyxl Êù•ËØªÂèñ .xlsx Êñá‰ª∂Ôºöpip install openpyxl")
    
    for file_path in all_files:
        file_name = os.path.basename(file_path)
        
        # Ê£ÄÊü•Êñá‰ª∂Êâ©Â±ïÂêç
        if file_name.endswith('.xlsx'):
            print(f"Ê≠£Âú®Âä†ËΩΩ Excel Êñá‰ª∂: {file_name}")
            try:
                # ËØªÂèñ Excel Êñá‰ª∂ÔºåÈªòËÆ§ËØªÂèñÁ¨¨‰∏Ä‰∏™ Sheet
                df = pd.read_excel(file_path)
                data_frames[file_name] = df
                print(f"  --> Âä†ËΩΩÊàêÂäüÔºåÂΩ¢Áä∂: {df.shape}")
            except Exception as e:
                print(f"  --> ‚ùå Excel Êñá‰ª∂ {file_name} Âä†ËΩΩÂ§±Ë¥•: {e}")

        elif file_name.endswith('.txt'):
            print(f"Ê≠£Âú®Âä†ËΩΩÊñáÊú¨Êñá‰ª∂: {file_name}")
            try:
                # ËØªÂèñÊñáÊú¨Êñá‰ª∂Ôºå‰ΩøÁî® 'utf-8' ÁºñÁ†Å
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                    text_contents[file_name] = content
                print(f"  --> Âä†ËΩΩÊàêÂäüÔºåÈïøÂ∫¶: {len(content)} Â≠óÁ¨¶")
            except Exception as e:
                print(f"  --> ‚ùå ÊñáÊú¨Êñá‰ª∂ {file_name} Âä†ËΩΩÂ§±Ë¥•: {e}")
        
        # ÂøΩÁï•ÂÖ∂‰ªñÊñá‰ª∂ (Â¶Ç .pdf, .gitignore Á≠â)
        else:
             print(f"Ë∑≥ËøáÊñá‰ª∂: {file_name}")

    print("\n====================================")
    print("‚úÖ ÊâÄÊúâÂèØÂä†ËΩΩÊñá‰ª∂Â§ÑÁêÜÂÆåÊØï„ÄÇ")
    print(f"DataFrame Êï∞Èáè: {len(data_frames)}")
    print(f"ÊñáÊú¨Êñá‰ª∂Êï∞Èáè: {len(text_contents)}")
    print("====================================")
    
    # Á§∫‰æãÔºöËÆøÈóÆÂÖ∂‰∏≠‰∏Ä‰∏™Êï∞ÊçÆË°®
    if data_frames:
        first_df_key = list(data_frames.keys())[0]
        print(f"\nÁ§∫‰æãÊï∞ÊçÆ (Êñá‰ª∂: {first_df_key}):")
        # ÊâìÂç∞ÂàóÂêçÂíåÂâç‰∏§Ë°å
        print(data_frames[first_df_key].head(2))

except Exception as e:
    print(f"ÂèëÁîüËá¥ÂëΩÈîôËØØ: {e}")
    print("ËØ∑Ê£ÄÊü• Python Â∫ìÊòØÂê¶ÂÆâË£ÖÂÆåÊï¥Ôºöpip install pandas openpyxl")
```

```python
import pandas as pd
import numpy as np
from IPython.display import display

# ==============================================================================
# ‚ö†Ô∏è Á§∫‰æãÊï∞ÊçÆËÆæÁΩÆ (ËØ∑Áî®ÊÇ®ÂÆûÈôÖÂä†ËΩΩÊï∞ÊçÆÁöÑ‰ª£Á†ÅÊõøÊç¢ËøôÈÉ®ÂàÜ)
#    ËøôÈáå‰ΩøÁî®ËôöÊãüÊï∞ÊçÆÊù•Ê®°ÊãüÊ∏ÖÊ¥óËøáÁ®ãÔºå‰ª•Á°Æ‰øù‰ª£Á†ÅÂèØÊâßË°åÂπ∂Â±ïÁ§∫ÊïàÊûú„ÄÇ
# ==============================================================================
data_frames = {
    'tariff_database_202405.xlsx': pd.DataFrame({
        'hts8': ['01012100', '01012900', '99038803', '99038804'],
        'mfn_ad_val_rate': [0.0, 0.017, 9999.99, 0.05], # Âç†‰ΩçÁ¨¶
        'brief_description': ['Horses', 'Donkeys', 'Tariff Text', 'Tariff Text'],
        'mexico_ad_val_rate': [0.0, 0.0, 10000.0, 0.0] # Âç†‰ΩçÁ¨¶
    }),
    'tariff_database_2025.xlsx': pd.DataFrame({
        'hts8': ['01012100', '01012900', '99038803'],
        'mfn_ad_val_rate': [0.02, 0.022, 9999.99], # Âç†‰ΩçÁ¨¶
        'brief_description': ['Horses 2', 'Donkeys 2', 'Tariff Text 2'],
        'mexico_ad_val_rate': [0.0, 0.0, 0.50]
    })
}

# Â≠òÂÇ®Ê∏ÖÊ¥óÂêéÁöÑ DataFrame
cleaned_data_frames = {}

PLACEHOLDER_THRESHOLD = 9999.0
print("=" * 80)
print(" " * 28 + "üßπ ÂºÄÂßãÊï∞ÊçÆÊ∏ÖÊ¥óÔºöÂç†‰ΩçÁ¨¶Â§ÑÁêÜ üßπ")
print("=" * 80)

# ÈÅçÂéÜÊâÄÊúâÊï∞ÊçÆË°®Âπ∂ÊâßË°åÊ∏ÖÊ¥ó
for file_name, df in data_frames.items():
    
    # Â§çÂà∂ DataFrame ‰ª•ËøõË°åÊ∏ÖÊ¥óÔºå‰øùÁïôÂéüÂßãÊï∞ÊçÆ
    df_cleaned = df.copy()
    
    print(f"\n\n{'#'*20} üßº Ê∏ÖÊ¥óÊï∞ÊçÆÈõÜÔºö{file_name} {'#'*20}")
    
    # 1. ËØÜÂà´ÊâÄÊúâÊï∞ÂÄºÂûãÂàó
    numeric_cols = df_cleaned.select_dtypes(include=['number']).columns
    
    # 2. Ê†∏ÂøÉÊ∏ÖÊ¥óÈÄªËæëÔºöÊõøÊç¢Âç†‰ΩçÁ¨¶
    for col in numeric_cols:
        # Â∞ÜÊâÄÊúâÂ§ß‰∫éÁ≠â‰∫éÈòàÂÄºÁöÑÂÄºÊõøÊç¢‰∏∫ NaN
        rows_replaced = df_cleaned.loc[df_cleaned[col] >= PLACEHOLDER_THRESHOLD, col].count()
        if rows_replaced > 0:
            df_cleaned.loc[df_cleaned[col] >= PLACEHOLDER_THRESHOLD, col] = np.nan
            print(f"   ‚ö´ Âàó '{col}' Â∑≤ÊõøÊç¢ {rows_replaced} ‰∏™Âç†‰ΩçÁ¨¶ (>={PLACEHOLDER_THRESHOLD}) ‰∏∫ NaN„ÄÇ")
            
    # 3. Â≠òÂÇ®Ê∏ÖÊ¥óÂêéÁöÑÊï∞ÊçÆ
    cleaned_data_frames[file_name] = df_cleaned
    print("   ‚úÖ Êï∞ÊçÆÊ∏ÖÊ¥óÂÆåÊàêÂπ∂Â∑≤Â≠òÂÇ®„ÄÇ")
    
    # 4. Â±ïÁ§∫Ê∏ÖÊ¥óÊïàÊûúÔºöÈáçÊñ∞ÊèèËø∞ÂÖ≥ÈîÆ‰ªé‰ª∑Á®éÁéáÂàó
    ad_val_rate_cols = [col for col in numeric_cols if '_ad_val_rate' in col]
    
    print(f"\n[-- Ê∏ÖÊ¥óÂêéÂÖ≥ÈîÆÂàó ({ad_val_rate_cols}) ÁªüËÆ°ÊèèËø∞ --]")
    
    with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.float_format', '{:.4f}'.format):
        try:
            display(df_cleaned[ad_val_rate_cols].describe())
        except Exception:
            print("‚ö†Ô∏è Ë≠¶ÂëäÔºöÊó†Ê≥ïÁîüÊàêÊñ∞ÁöÑÁªüËÆ°ÊèèËø∞„ÄÇ")
    
    print(f"**" * 40 + "\n")

print("=" * 80)
print(" " * 25 + "üéâ ÊâÄÊúâÊï∞ÊçÆÁöÑÂç†‰ΩçÁ¨¶Ê∏ÖÊ¥óÂ∑≤ÂÆåÊàêÔºÅ")
print("=" * 80)

# ÊâìÂç∞ÊúÄÁªàÊèêÁ§∫
print("\n[ÂêéÁª≠Ê≠•È™§ÊèêÁ§∫]")
print("ÊÇ®Áé∞Âú®ÂèØ‰ª•‰ΩøÁî® 'cleaned_data_frames' Â≠óÂÖ∏‰∏≠ÁöÑÊï∞ÊçÆËøõË°åËøõ‰∏ÄÊ≠•ÁöÑÂàÜÊûêÔºå‰æãÂ¶ÇÔºö")
print("1. ÊØîËæÉ‰∏çÂêåÂπ¥‰ªΩÁöÑÂπ≥ÂùáÁ®éÁéáÂíåÁ®éÁéáÂàÜÂ∏É„ÄÇ")
print("2. Êé¢Á¥¢ HTS ÁºñÁ†ÅÁöÑÁº∫Â§±ÂÄºÊàñÈùûÊï∞ÂÄºÂûãÂàó„ÄÇ")
```

```python
import pandas as pd
import numpy as np
from IPython.display import display

# ==============================================================================
# I. ÂÅáËÆæÂíåÂèÇÊï∞ËÆæÂÆö (Âü∫‰∫éÂ§ñÈÉ®Ë¥∏ÊòìÊï∞ÊçÆÂíåÁªèÊµéÂºπÊÄßÁêÜËÆ∫ÁöÑÁÆÄÂåñÊ®°Âûã)
# ==============================================================================

# HTS ÁºñÁ†Å (‰ªÖÁî®‰∫éÂèÇËÄÉÂíåËØ¥Êòé)
SOYBEAN_HTS_CODES = ['1201', '2304']

# ÂÖ≥Á®éË∞ÉÊï¥Ââç (2024Âπ¥) ÁöÑË¥∏ÊòìÊï∞ÊçÆÂíåÁªèÊµéÂèÇÊï∞ (ÂÅáËÆæÂÄº)
ASSUMED_TRADE_DATA = {
    'USA': {
        'Q0_million_tons': 25.0,     # ÂàùÂßãÂØπÂçéÂá∫Âè£Èáè (Áôæ‰∏áÂê®)
        'Price_per_ton': 600,        # Âá∫Âè£Âπ≥Âùá‰ª∑Ê†º ($/Âê®)
        'Tariff_Change_Delta_t': 0.25, # ‰∏≠ÂõΩÂØπÁæéÊÉ©ÁΩöÊÄßÂÖ≥Á®éÂ¢ûÂä† 25% (Œît)
        'Trade_Loss_Share': 1.0,     # 100% ÁöÑË¥∏ÊòìÊçüÂ§±Áî®‰∫éËÆ°ÁÆóË¥∏ÊòìËΩ¨Áßª
    },
    'BRA': {
        'Q0_million_tons': 50.0,
        'Price_per_ton': 600,
        'Tariff_Change_Delta_t': 0.0, # Â∑¥Ë•ø/ÈòøÊ†πÂª∑Èù¢‰∏¥ÁöÑÂÖ≥Á®é‰∏çÂèò
        'Trade_Loss_Share': 0.8,     # ÂÅáËÆæËé∑ÂæóÁæéÂõΩÊçüÂ§±‰ªΩÈ¢ùÁöÑ 80%
    },
    'ARG': {
        'Q0_million_tons': 5.0,
        'Price_per_ton': 600,
        'Tariff_Change_Delta_t': 0.0,
        'Trade_Loss_Share': 0.2,     # ÂÅáËÆæËé∑ÂæóÁæéÂõΩÊçüÂ§±‰ªΩÈ¢ùÁöÑ 20%
    }
}

# ÁªèÊµéÂºπÊÄßÂèÇÊï∞
DEMAND_ELASTICITY = -0.5  # ‰∏≠ÂõΩÂØπÂ§ßË±ÜÁöÑËøõÂè£ÈúÄÊ±ÇÂºπÊÄß (‰∏≠Â∫¶‰∏çÂºπÊÄß)
INITIAL_TARIFF = 0.0      # ÂàùÂßãÂü∫Á°ÄÂÖ≥Á®é (t0) ÂÅáËÆæ‰∏∫ 0%

# ==============================================================================
# II. ËØÑ‰º∞Ê®°ÂûãÂáΩÊï∞ÔºöÂü∫‰∫éÂºπÊÄßËÆ°ÁÆóÂá∫Âè£ÈáèÂèòÂåñ
# ==============================================================================

def calculate_tariff_impact(Q0, elasticity, delta_t, t0):
    """
    ‰ΩøÁî®ÂºπÊÄßÂÖ¨Âºè‰º∞ÁÆóÂÖ≥Á®éÂèòÂåñ (Œît) ÂØπÂá∫Âè£Èáè (Q) ÁöÑÂΩ±Âìç (ŒîQ)„ÄÇ
    ŒîQ = Q0 * ŒµD * (Œît / (1 + t0))
    """
    if delta_t == 0:
        return 0.0  # ÂÖ≥Á®é‰∏çÂèòÔºåÂá∫Âè£ÈáèÂèòÂåñ‰∏∫ 0
    
    # ËÆ°ÁÆó‰ª∑Ê†ºÂèòÂåñÁôæÂàÜÊØî
    price_change_ratio = delta_t / (1 + t0)
    
    # ËÆ°ÁÆóÂá∫Âè£ÈáèÂèòÂåñ (ŒîQ)
    delta_Q = Q0 * elasticity * price_change_ratio
    return delta_Q

# ==============================================================================
# III. Ê®°ÂûãËÆ°ÁÆó‰∏éÁªìÊûúËæìÂá∫
# ==============================================================================

results = []
us_trade_loss = 0.0 # Áî®‰∫éÂ≠òÂÇ®ÁæéÂõΩÊçüÂ§±ÁöÑË¥∏ÊòìÈáèÔºå‰ª•‰æõÂ∑¥Ë•øÂíåÈòøÊ†πÂª∑ÂàÜÈÖç

print("=" * 80)
print(" " * 15 + "üìä ËØÑ‰º∞Ê®°ÂûãËÆ°ÁÆóÔºöÁæéÂõΩÂÖ≥Á®éË∞ÉÊï¥ÂØπÂ§ßË±ÜÂá∫Âè£ÁöÑÂΩ±Âìç")
print("=" * 80)

# 1. ËÆ°ÁÆóÁæéÂõΩÂá∫Âè£ÂèòÂåñ
country = 'USA'
data = ASSUMED_TRADE_DATA[country]
Q0 = data['Q0_million_tons']
P = data['Price_per_ton']
delta_t = data['Tariff_Change_Delta_t']

# ËÆ°ÁÆóÁæéÂõΩÂá∫Âè£ÈáèÂèòÂåñ
delta_Q = calculate_tariff_impact(Q0, DEMAND_ELASTICITY, delta_t, INITIAL_TARIFF)

# Â≠òÂÇ®ÁæéÂõΩÊçüÂ§±ÁöÑË¥∏ÊòìÈáè
us_trade_loss = abs(delta_Q)

Q1 = Q0 + delta_Q
V1_billion = (Q1 * P) / 1_000_000  # ËΩ¨Êç¢‰∏∫‰∫øÁæéÂÖÉ

results.append({
    'ÂõΩÂÆ∂': country,
    'ÂàùÂßãÂá∫Âè£Èáè (‰∏áÂê®)': Q0,
    'ÂÖ≥Á®éÂèòÂåñ (Œît)': f'{delta_t*100:.0f}%',
    'Âá∫Âè£ÈáèÂèòÂåñ (ŒîQ)': delta_Q,
    'Ë∞ÉÊï¥ÂêéÂá∫Âè£Èáè (‰∏áÂê®)': Q1,
    'Ë∞ÉÊï¥ÂêéÂá∫Âè£‰ª∑ÂÄº (‰∫øÁæéÂÖÉ)': V1_billion
})
print(f"‚ö´ ÁæéÂõΩ (USA) ÂÖ≥Á®éÂΩ±ÂìçËÆ°ÁÆóÂÆåÊàê„ÄÇÊçüÂ§±Ë¥∏ÊòìÈáè: {us_trade_loss:.2f} Áôæ‰∏áÂê®„ÄÇ")


# 2. ËÆ°ÁÆóË¥∏ÊòìËΩ¨ÁßªÂèóÁõäËÄÖÔºöÂ∑¥Ë•øÂíåÈòøÊ†πÂª∑
for country in ['BRA', 'ARG']:
    data = ASSUMED_TRADE_DATA[country]
    Q0 = data['Q0_million_tons']
    P = data['Price_per_ton']
    loss_share = data['Trade_Loss_Share']

    # Ë¥∏ÊòìËΩ¨ÁßªËé∑ÂæóÁöÑÂ¢ûÈáè
    delta_Q_gain = us_trade_loss * loss_share

    Q1 = Q0 + delta_Q_gain
    V1_billion = (Q1 * P) / 1_000_000

    results.append({
        'ÂõΩÂÆ∂': country,
        'ÂàùÂßãÂá∫Âè£Èáè (‰∏áÂê®)': Q0,
        'ÂÖ≥Á®éÂèòÂåñ (Œît)': '0.0%',
        'Âá∫Âè£ÈáèÂèòÂåñ (ŒîQ)': delta_Q_gain,
        'Ë∞ÉÊï¥ÂêéÂá∫Âè£Èáè (‰∏áÂê®)': Q1,
        'Ë∞ÉÊï¥ÂêéÂá∫Âè£‰ª∑ÂÄº (‰∫øÁæéÂÖÉ)': V1_billion
    })
    print(f"‚ö´ {country} Ë¥∏ÊòìËΩ¨ÁßªÂàÜÈÖçÂÆåÊàê„ÄÇËé∑ÂæóË¥∏ÊòìÈáè: {delta_Q_gain:.2f} Áôæ‰∏áÂê®„ÄÇ")


# 3. ËæìÂá∫ÊúÄÁªàÁªìÊûú
df_results = pd.DataFrame(results)

print("\n--- üìà Ë∞ÉÊï¥ÂêéÁöÑ‰∏âÂõΩÂ§ßË±ÜÂá∫Âè£ÈáèÂíåÂá∫Âè£‰ª∑ÂÄºÂàÜÂ∏É‰º∞ËÆ° ---")
# ËÆæÁΩÆÊòæÁ§∫Ê†ºÂºè
pd.set_option('display.float_format', '{:.2f}'.format)
display(df_results)
print("\n*Ê≥®: ÁªìÊûúÂü∫‰∫é‰∏≠ÂõΩÂØπÁæéÂõΩÂ§ßË±ÜÂæÅÊî∂ 25% ÊÉ©ÁΩöÊÄßÂÖ≥Á®éÁöÑË¥∏ÊòìËΩ¨ÁßªÊÉÖÊôØÂíåÂÅáËÆæÁöÑÁªèÊµéÂºπÊÄß (-0.5)„ÄÇ")
print("=" * 80)
```
